```python
  import numpy as np
  import pandas as pd
  import matplotlib.pyplot as plt
  import matplotlib.image as mpimg
  
  from scipy.stats import multivariate_normal as mvn
  
  %matplotlib inline
  df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MNIST_train.csv')
  df.head()
```

```python
df.shape
```

```python
df_array = df.to_numpy() #turn the training dataset into an array
df_array
```

```python
print_NaN_rows = df[df.isnull().any(axis=1)] #list the rows where NaN appears
print_NaN_rows
```

```python
print_None_rows = df[df.eq("None").any(axis=1)]  #check to see if there are any 'None' values in any row
print_NaN_rows.shape
```

```python
y = df_array[:,2] #extract the label column
y
```

```python
y.shape
```

```python
X = np.delete(df_array, [0,1,2], axis=1)
```

```python
X_normalised = X/255 
```

```python
#using Naive_Bayes
class GaussNB():

  def fit(self, X,y, epsilon):
    self.likelihoods = dict()
    self.priors = dict()

    self.K = set(y.astype(int))

    for k in self.K:
      X_k = X[y==k,:]
      self.likelihoods[k] = {"mean":X_k.mean(axis=0), "cov":X_k.var(axis=0) + epsilon}
      self.priors[k]= len(X_k)/len(X)
  

  def predict(self, X):
    N, D = X.shape

    P_hat = np.zeros((N,len(self.K)))

    for k, l in self.likelihoods.items():
      P_hat[:,k] = mvn.logpdf(X, l["mean"], l["cov"]) + np.log(self.priors[k])
    
    return P_hat.argmax(axis=1)
```

```python
gnb = GaussNB()
```

```python
def accuracy_training_GNB(y,y_hat_training_GNB):
  return np.mean(y== y_hat_training_GNB)

result_GNB_training = []
iteration = []

for i in np.arange(0.001,0.5,0.001):
  gnb.fit(X_normalised,y,i)
  y_hat_training_GNB=gnb.predict(X_normalised)
  accuracy = accuracy_training_GNB(y,y_hat_training_GNB)
  result_GNB_training.append(accuracy) 
  iteration.append(i)
  print("i =", i, "accuracy =", accuracy) #training accuracy
result_GNB_df = pd.DataFrame(result_GNB_training)
iteration_df = pd.DataFrame(iteration)
result_GNB_df.to_csv('result_gnb_train.csv')
iteration_df.to_csv('iteration_train.csv')
```
