```python
X = np.delete(df_array, [0,1,2], axis=1)
```

```python
X_normalised = X/255 
```

```python
class GaussBayes():

  def fit(self, X, y, epsilon):
    self.likelihoods =dict()
    self.priors = dict()

    self.K = set(y.astype(int))

    for k in self.K:
      X_k = X[y==k,:]
      N_k, D = X_k.shape
      mu_k = X_k.mean(axis=0)
      self.likelihoods[k] ={"mean":X_k.mean(axis=0),"cov":(1/(N_k-1))*np.matmul((X_k-mu_k).T,X_k-mu_k)+ epsilon*np.identity(D)}
      self.priors[k] = len(X_k)/len(X)


  def predict(self, X):
    N, D = X.shape

    P_hat = np.zeros((N,len(self.K)))

    for k, l in self.likelihoods.items():
      P_hat[:,k] = mvn.logpdf(X,l["mean"],l["cov"]) + np.log(self.priors[k])
    return P_hat.argmax(axis=1)
```

```python
gbayes= GaussBayes()
```

```python
def accuracy_training_GB(y,y_hat_training_gbayes):
  return np.mean(y== y_hat_training_gbayes)

result_gbayes_training = []
iteration = []

for i in np.arange(0.001,0.4,0.001):
  gbayes.fit(X_normalised,y,i)
  y_hat_training_gbayes=gbayes.predict(X_normalised)
  accuracy = accuracy_training_GB(y,y_hat_training_gbayes) #training accuracy
  result_gbayes_training.append(accuracy)
  iteration.append(i)
  print("i =", i, "accuracy =", accuracy) #training accuracy
result_GB_df = pd.DataFrame(result_gbayes_training)
iteration_df = pd.DataFrame(iteration)
result_GNB_df.to_csv('result_gbayes_train.csv')
iteration_df.to_csv('iteration_gbayes_train.csv')
```
