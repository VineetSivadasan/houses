```python
  import numpy as np
  import pandas as pd
  import matplotlib.pyplot as plt
  import matplotlib.image as mpimg
  
  from scipy.stats import multivariate_normal as mvn
  
  %matplotlib inline
  df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MNIST_train.csv')
  df.head()
```

```python
df_array = df.to_numpy() #turn the training dataset into an array
df_array
```

```python
print_NaN_rows = df[df.isnull().any(axis=1)] #list the rows where NaN appears
print_NaN_rows
```

```python
print_None_rows = df[df.eq("None").any(axis=1)]  #check to see if there are any 'None' values in any row
print_NaN_rows.shape
```

```python
y = df_array[:,2] #extract the label column
y
```

```python
y.shape
```

```python
X = np.delete(df_array, [0,1,2], axis=1)
```

```python
X_normalised = X/255
```

```python
#using KNN method
class KNNClassifier():
  def fit(self,X,y):
    self.X=X
    self.y=y

  def predict(self, X, K, epsilon=1e-3):
    N=len(X)
    y_hat_KNN =  np.zeros(N)

    for i in range (N):
      dist2=np.sum((self.X-X[i])**2, axis=1)
      idxt = np.argsort(dist2)[:K]
      gamma_k = 1/(np.sqrt(dist2[idxt])+epsilon)
      y_hat_KNN[i] =np.bincount(self.y[idxt], weights=gamma_k).argmax()
    
    return y_hat_KNN
    
```

```python
knn=KNNClassifier()
knn.fit(X_normalised,y)
```

```python
y_hat_KNN = knn.predict(X_normalised,20)
```

```python
def accuracy(y,y_hat_KNN):
  return np.mean(y== y_hat_KNN)


accuracy(y,y_hat_KNN)
```
